```python
#-<general modules>-#
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
from astropy.io import fits
import glob
```

```python
noiseadd_ext = "_noise_added"
rnoise_ext   = "_real_noise_stamp"
spoiss_ext   = "_smoothed_poiss"
```

```python
image_list=glob.glob("VELA01/VELA01*.fits")

one_image = 3
image = image_list[one_image].split(".fits")[0]
image = "VELA01/VELA01_a0.060_sunrise_cam00_WFC3-F160W" 

imagename = image.split(".fits")[0]  ; print(imagename)
hdul_noiseadd = fits.open( imagename + noiseadd_ext + '.fits')
hdul_rnoise   = fits.open( imagename + rnoise_ext   + '.fits')
hdul_spoiss   = fits.open( imagename + rnoise_ext   + '.fits')


```

```python
hdul_noiseadd.info()
hdul_rnoise.info()
hdul_spoiss.info()
```

```python
data_noiseadd = hdul_noiseadd[0].data
data_rnoise = hdul_rnoise[0].data
data_spoiss = hdul_spoiss[0].data
```

```python
def plot_noisegen(one_image = 3):
    image = image_list[one_image].split(".fits")[0]
    
    fig,(ax1,ax2,ax3) = plt.subplots(1,3)
    with fits.open(imagename + noiseadd_ext + '.fits') as hdul_noiseadd :  
        data_noiseadd = hdul_noiseadd[0].data
        ax1.imshow(data_noiseadd) 
    with fits.open( imagename + rnoise_ext   + '.fits') as hdul_rnoise :
        data_rnoise = hdul_rnoise[0].data
        ax2.imshow(data_rnoise) 
    with fits.open( imagename + rnoise_ext   + '.fits') as hdul_spoiss:
        data_spoiss = hdul_spoiss[0].data
        ax3.imshow(data_spoiss)
    

    
```

```python
plot_noisegen(5)
```
