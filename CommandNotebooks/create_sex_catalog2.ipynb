```python
###---<MAIN LOOP>---### : Adapted for Python3.4
try:                 #reset dataframe df
    del(df)
except:
    pass
df = pd.DataFrame() 
i = 0
for f in files:
    
    if (i % dot_freq == 0):                        #<ME> con cierta frecuencia imprime un mensaje
        sys.stdout.write(str(i/dot_freq)+"%")
        sys.stdout.flush()
           
    seg_name = f.split('.fits')[0]+'_seg.tmp.fits'         #<ME> filename+"_seg.tmp.fits"   #CHECKIMAGE_NAME SEGMENTATION    
    cat_name = f.split('.fits')[0]+'_cat.tmp.csv'          #<ME> filename+"_cat.tmp.csv"    #CATALOG_TYPE ASCII_HEAD

    #<ME> cmd: sextractor -c /notebooks/CLUMPS_VELA/sextractor_work/clumps.sex %s -CHECKIMAGE_NAME %s -CATALOG_NAME %s
    sub = subprocess.Popen(cmd % (f,seg_name,cat_name), shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)  
    #<ME> %s(1): f   ;; %s(2): seg_name ;; %s(3): cat_name
    exit_code = sub.wait()
    stdout, stderr = sub.communicate()
    '''
    #SYNTAX: sextractor <image> [<image2>][-c <configuration_file>][-<keyword> <value>]
     class subprocess.Popen(args, bufsize=0, executable=None, stdin=None, stdout=None, stderr=None, 
     preexec_fn=None, close_fds=False, shell=False, cwd=None, env=None, universal_newlines=False, startupinfo=None, creationflags=0)
    
    -CHECKIMAGE_NAME in "clumps.sex": CHECKIMAGE_NAME me_check.fits (overwrited by option)
     "":CATALOG_TYPE ASCII_HEAD
    '''
    if (exit_code != 0):                                  #<ME> Error print message
        print(cmd % (f,seg_name,cat_name))
        print("Non zero exit code... that's a problem!")
        continue                                          #<ME> Continue to next file
        
    seg        = fits.open(seg_name)[0].data
    
    seg_arr[i] = seg
    cat = pd.read_csv(cat_name,delim_whitespace=True,comment='#',names=cols)
    #<ME> delim_whitespace=True: nada se trata como espacio en blanco; 
    #comment: linea de comentario empezando por "str"
    #names: lista de columna de nombres que usa
    # cat is now an empty dataframe with those columns
    # len(cat) is the number of clumps detected in img
    #</ME>
    #---------------------------DATAFRAME_COLUMNS----------------------------#
    gal_id = f.split("/")[-1].split("_")[0]     #; print(gal_id) #i.e. VELA06
    #<ME> f.split("/")[-1] se queda con el nombre de la imagen y la extensión, 
    #.split(".fits")[0] a continuación se queda sólo con el nombre,
    #.split("-")[-2] se queda con instrumento pero sin filtro #</ME>
    cat['gal_id']    = gal_id  #;print(gal_id)#<ME> creates a new col called "gal_id"
    cat['seg_index'] = i       #;print(cat['seg_index'])
    
    
    #--<ME>--#
    a0 = f.split("/")[-1].split("_")[1].split(".")[1]      #; print(a0)     #; a0
    cat['a0'] = "%.2f"%(float(a0)/100.)
    
    cam = f.split("/")[-1].split("_")[3].split("cam")[-1]  #; print(cam)    #; cam   = "cam00"
    cat['cam'] = cam
    
    instr = f.split("/")[-1].split("_")[4].split("-")[0]  #;print(instr)  #; instr= "ACS"
    cat['instrument'] = instr
    
    filt = f.split("/")[-1].split("-")[1].split("_")[0]   #;print(filt)   #; filt = "F435W"
    cat['filter'] = filt
    
    filename = f.split("/")[-1].split(".fits")[0] 
    cat['filename'] = filename                                #;print(filename)#; filename = "..."   
 
    #--</ME>--#
    # Initialize varibles for clumps detected in catalog csv file
    prob_mean = np.empty(len(cat))       
    prob_int  = np.empty(len(cat))
    flux      = np.empty(len(cat))
    
    # Open data of real_imgs and pred_imgs
    pred = fits.open(f)[0].data                                             #<ME> Imagen de la prediccion (DEFAULT)
    img  = fits.open(f.split('-pred.fits')[0]+'.fits')[0].data              #<ME> Imagen que acaba en .fits. Estas imagenes y las "-pred" en la misma carpeta
    #pdb.set_trace()
    
    for j in range(len(cat)):
        pixels = np.where(seg == (j+1))
        if (len(pixels[0]) == 0):      
            continue
        #print("No matching pixels in seg image for i=%d, clump=%d (tot: %d)"%(i,j,len(cat)))    
        #print(np.max(seg))
        prob_mean[j] = np.mean(pred[pixels])
        prob_int[j]  = np.sum(pred[pixels])
        flux[j]      = np.sum(img[pixels])
        
    cat['prob_mean'] = prob_mean
    cat['prob_int']  = prob_int
    cat['flux']      = flux
    
    df = df.append(cat)   # Appends the new builded "cat"(of 1 file) to "df"
    #pdb.set_trace()      # Debug
    
    # remove SExtractor output files we just read in
    #'''
    try:                                      #<ME> intenta remover los archivos si existen en el dir
        os.remove(seg_name)                   #% -CHECKIMAGE_NAME           
        os.remove(cat_name)                   #% -CATALOG_NAME
    except:
        print("Error in removing 'seg_name' and/or 'cat_name'")
        pass    
    #'''
    i += 1

print("Done!")
```
