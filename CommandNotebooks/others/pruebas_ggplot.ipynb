```python
###---------<pruebas_ggplot.ipynb>---------###
#!pip install ggplot2
'''
ggplot is based on ggplot2, an R plotting system, and concepts from The Grammar of Graphics. ggplot operates differently than matplotlib: 
it lets you layer components to create a complete plot. For instance, you can start with axes, then add points, then a line, a trendline, etc. 
Although The Grammar of Graphics has been praised as an “intuitive” method for plotting, seasoned matplotlib users might need time to adjust to this new mindset.

According to the creator, ggplot isn't designed for creating highly customized graphics. It sacrifices complexity for a simpler method of plotting.

ggplot is tightly integrated with pandas, so it's best to store your data in a DataFrame when using ggplot. 
'''
```

```python
from ggplot import *
```

```python
ggplot(aes(x='date', y='beef'), data=meat) +  geom_line() + stat_smooth(colour='blue', span=0.2)
```

```python
ggplot(diamonds, aes(x='carat', y='price', color='cut')) +\
    geom_point() +\
    scale_color_brewer(type='diverging', palette=4) +\
    xlab("Carats") + ylab("Price") + ggtitle("Diamonds")
```

```python
ggplot(diamonds, aes(x='price', fill='cut')) +\
    geom_density(alpha=0.25) +\
    facet_wrap("clarity")
```
